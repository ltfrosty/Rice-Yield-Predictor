```python
master_df: pd.DataFrame = pd.read_csv("../Raw Data/Population Stats.csv")
#print(master_df)
```

```python
working_df = deepcopy(master_df)
```

```python
#rename column names
working_df.rename(columns={"Area": "Country"}, inplace = True)
#print(working_df)
```

```python
columns_of_interest = ["Country", "Element", "Year", "Value"]
working_df = working_df[columns_of_interest]
#print(working_df)
```

```python
working_df_dated = working_df[(working_df.Year >= 1990) & (working_df.Year <= 2019)]
#print(working_df_dated)

countries = ["China", "India", "Thailand", "Viet Nam"]
working_df_dated_country = working_df_dated[working_df_dated.Country.isin(countries)]
#print(working_df_dated_country)
```

## India dataframes

```python
###Segment out the india data
india_df = working_df_dated_country[working_df_dated_country.Country == "India"]
#print(india_df)

#get total pop
india_total_pop_df = india_df[india_df.Element == "Total Population - Both sexes"]
india_total_pop_array = india_total_pop_df["Value"].to_numpy()
#india_total_pop_df.reset_index(drop = True, inplace = True)
#print(india_total_pop_array)

india_rural_pop_df = india_df[india_df.Element == "Rural population"]
india_rural_pop_array = india_rural_pop_df["Value"].to_numpy()
#india_rural_pop_df.reset_index(drop = True, inplace = True)
#print(india_rural_pop_array)

india_urban_pop_df = india_df[india_df.Element == "Urban population"]
india_urban_pop_array = india_urban_pop_df["Value"].to_numpy()

#india_urban_pop_df.reset_index(drop = True, inplace = True)
#print(india_urban_pop_array)

###Here are the 2 india data frames we want to model.
india_rural_over_total_array = india_rural_pop_array / india_total_pop_array
india_rural_over_total_df = pd.DataFrame(india_rural_over_total_array, columns=["india rural/total"])

india_urban_over_total_array = india_urban_pop_array / india_total_pop_array
india_urban_over_total_df = pd.DataFrame(india_urban_over_total_array, columns=["india urban/total"])

#print(india_rural_over_total_df)
#print(india_urban_over_total_df)
```
